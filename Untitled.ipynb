```python
import pandas as pd
import numpy as np
from IPython.display import display 
```

```python
data = pd.read_excel('results.xlsx') # all the records after the join
display(data.head(10))

data=pd.DataFrame(data)
```

```python
data1 = data[data.city=='New York'] # removing all the data of other cities
```

```python
import matplotlib.pyplot as plt # package that helps in plotting
```

### Calculating moving average for both global and city temperatures

```python
global_mv_avg = data1['global_avg_temp'].rolling(7).mean()  # rolling(7) because we are considering 7-year MA
city_mv_avg = data1['city_avg_temp'].rolling(7).mean()
```

```python
display(city_mv_avg.head(10))
```

### Comparing the data using line chart

```python
plt.plot(data1['year'], city_mv_avg, label = 'New York')
plt.plot(data1['year'], global_mv_avg, label='Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature < in Celcius >")
plt.title("Temperature comparison <of New York and Global average>")
plt.show()
```
